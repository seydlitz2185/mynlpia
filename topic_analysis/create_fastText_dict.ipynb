```python
import fasttext
import fasttext.util
import pandas as pd
import numpy as np
import os
ft = fasttext.load_model('cc.zh.300.bin')
ft.get_dimension()
```

```python
df = pd.read_csv('../dataset/data.csv')
docs  =df.seg_text
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import casual_tokenize
tfidf = TfidfVectorizer(use_idf=True, smooth_idf=True, norm=None)
tfidf_docs =tfidf.fit_transform(raw_documents=df.seg_text).toarray()
tfidf_docs = tfidf_docs - tfidf_docs.mean(axis=0)
words = tfidf.get_feature_names_out()
```

```python
def get_word_vector(word):
    return ft.get_word_vector(word)
```

```python
for i,word in enumerate(words):
    if i == 0:
        word_vectors = get_word_vector(word)
    else:
        word_vectors = np.vstack((word_vectors,get_word_vector(word)))
word_dict = dict(zip(words,word_vectors))
np.savez('../dataset/embedding_cc.zh.300.npz',embeddings = word_vectors) 
```
